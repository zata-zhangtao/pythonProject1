```python
import torch
learing_rate = 0.01
```

```python
#生成x和y数据
x = torch.rand([500,1])
y_true = x*3 + 0.8
```

```python
w = torch.rand([1,1],requires_grad=True)#rand 函数返回一个随机均匀分布的张量
print(w)
b = torch.tensor(0,requires_grad=True,dtype=torch.float32)
print(b)
```

```python
for i in range(10000):
    y_prridct = torch.matmul(x,w)+b
    loss = 0
    loss = (y_true - y_prridct).pow(2).mean()

    if w.grad is not None:
        w.grad.data.zero_()
    if b.grad is not None:
        b.grad.data.zero_()

    loss.backward()
    w.data = w.data - learing_rate*w.grad
    b.data = b.data - learing_rate*b.grad
    # print("w,b,loss",w.item(),b.item(),loss.item())
```
